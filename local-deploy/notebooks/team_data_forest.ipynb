# "Feature Importances"

```python
from sklearn import tree
import pandas as pd
import os
```

```python
data = pd.read_csv("derby_scored.csv")
data.head()
```

```python
data = data.drop("pos", axis=1 )
data = data.drop("timestamp", axis=1 )
data = data.drop("player", axis=1 )
data = data.drop("org", axis=1 )
data = data.drop("sport", axis=1 )
data = data.drop("tweet", axis=1 )
data = data.drop("plus", axis=1 )
#.dropna(axis=0, how='any', inplace = True)
# data.reset_index(drop=True)
data
```

```python
target = data["value"]
target_names = ["positive", "negative"]
```

```python
data = data.drop("value", axis=1)
feature_names = data.columns
data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

# Roller Derby Most Contributing to Red Flags


```python
factors = sorted(zip(rf.feature_importances_, feature_names), reverse=True)
factors = pd.DataFrame(factors)
factors
```
